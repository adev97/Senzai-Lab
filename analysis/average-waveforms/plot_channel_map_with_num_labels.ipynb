```python
import os
import numpy as np
import matplotlib.pyplot as plt

## plot channel map with color coded channel numbers

kilo_sort_dir = r"D:\Kilosort\Mouse08_SC_20251007_810to2250\kilosort4"
os.chdir(kilo_sort_dir)

chanMap = np.load('channel_map.npy')
pos = np.load('channel_positions.npy')

x = pos[:, 0]
y = pos[:, 1]

plt.figure(figsize=(4,8))
plt.scatter(x, y, c=chanMap, s=20)
plt.gca().invert_yaxis()
plt.axis('equal')
plt.colorbar(label='Channel index (0-based)')
plt.xlabel('X (µm)')
plt.ylabel('Y (µm)')
plt.title('Kilosort4 Channel Map')
plt.show()
```

```python
## plot channel map with number labeled channels
plt.figure(figsize=(5,10))
plt.scatter(x, y, c='skyblue', s=40)  # channel points

# Invert Y to match probe depth
plt.gca().invert_yaxis()
plt.axis('equal')
plt.xlabel('X (µm)')
plt.ylabel('Y (µm)')
plt.title('Kilosort4 Channel Map')

# Add channel numbers
for i in range(len(chanMap)):
    plt.text(x[i]+2, y[i], str(chanMap[i]), fontsize=7, color='black')

plt.show()
```
